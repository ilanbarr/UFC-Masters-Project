```python
from datetime import datetime
def date_reformat(date_string):
    """
    Converts a date string from "Month DD, YYYY" format to "YYYY-MM-DD" format.
    
    Parameters:
    - date_string (str): The date in "Month DD, YYYY" format.
    
    Returns:
    - str: The reformatted date string in "YYYY-MM-DD" format.
    """
    # Parse the date string into a datetime object
    date_object = datetime.strptime(date_string, '%B %d, %Y')
    
    # Convert the datetime object back to a string in the desired format
    formatted_date_string = date_object.strftime('%Y-%m-%d')
    
    return formatted_date_string
df_fight_decision_data['Date'] = df_fight_decision_data['Date'].apply(lambda x: date_reformat(x))
```

```python
def get_judge_score(row, judge_number):
    # Determine if the row's 'Fighter' is Fighter 1 or Fighter 2
    if row['Fighter'] == row['Fighter 1']:
        fighter_position = 'Fighter 1'
    else:
        fighter_position = 'Fighter 2'
    
    # Construct the column name based on the current round, fighter position, and judge number
    column_name = f'Judge {judge_number} {fighter_position} Round {int(row["Round"])}'
    
    # Return the score from the corresponding column
    return row.get(column_name, None)  # Using .get() to avoid KeyError if column does not exist

# # Apply the function to create new columns for each judge's score
# df_merged['Judge 1 score'] = df_merged.apply(get_judge_score, judge_number=1, axis=1)
# df_merged['Judge 2 score'] = df_merged.apply(get_judge_score, judge_number=2, axis=1)
# df_merged['Judge 3 score'] = df_merged.apply(get_judge_score, judge_number=3, axis=1)

# # Display the modified DataFrame to verify the new columns
# df_merged[['Round', 'Fighter', 'Fighter 1', 'Fighter 2', 'Judge 1 score', 'Judge 2 score', 'Judge 3 score']].head()
```
