```python
import pandas as pd
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
import pickle

# Loading ROC data from a pickle file
with open('l2_log_reg_roc_data.pkl', 'rb') as file:
    roc_data_logistic = pickle.load(file)

fpr_logistic = roc_data_logistic['fpr']
tpr_logistic = roc_data_logistic['tpr']
roc_auc_logistic = roc_data_logistic['roc_auc']
```

```python
import pickle


with open('roc_data_betting.pkl', 'rb') as file:
    roc_data_betting = pickle.load(file)

# Assigning variables for the betting odds model
fpr_betting = roc_data_betting['fpr']
tpr_betting = roc_data_betting['tpr']
roc_auc_betting = roc_data_betting['roc_auc']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))  # Increased figure size for better visibility

# Plot ROC curve for the logistic regression model
plt.plot(fpr_logistic, tpr_logistic, color='blue', lw=2, label='ROC curve Logistic Model (area = %0.2f)' % roc_auc_logistic)

# Plot ROC curve for the betting odds model
plt.plot(fpr_betting, tpr_betting, color='green', lw=2, label='ROC curve Betting Odds Model (area = %0.2f)' % roc_auc_betting)

# Plot diagonal 50% line
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

# Customizing the plot with even larger fonts
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)  # Even larger font for X-axis label
plt.ylabel('True Positive Rate', fontsize=18)  # Even larger font for Y-axis label
plt.title('Comparison of ROC Curves', fontsize=20)  # Even larger font for title
plt.legend(loc="lower right", fontsize=16)  # Even larger font for legend

# Display the plot
plt.show()
```
